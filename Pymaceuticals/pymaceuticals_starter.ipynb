```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy import stats

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_data = pd.read_csv(mouse_drug_data_to_load)
clinical_trial_data = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
# merge both datasets 
mouse_clinicaltrail_df = pd.merge(mouse_drug_data, clinical_trial_data, on = "Mouse ID")

# Display the data table for preview
mouse_clinicaltrail_df
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
mean_tumor = pd.DataFrame(mouse_clinicaltrail_df.groupby(["Drug","Timepoint"])["Tumor Volume (mm3)"].mean())
mean_tumor
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
TVsem_df = pd.DataFrame(mouse_clinicaltrail_df.groupby(["Drug","Timepoint"])["Tumor Volume (mm3)"].sem())
TVsem_df.head()
# Convert to DataFrame

# Preview DataFrame

TVsemtrans = TVsem_df.pivot_table(columns='Drug', index='Timepoint', values='Tumor Volume (mm3)')
TVsemPivoted_df = pd.DataFrame(TVsemtrans)
TVsemPivoted_df
```

```python
# Minor Data Munging to Re-Format the Data Frames

TVtrans = mean_tumor.pivot_table(columns='Drug', index='Timepoint', values='Tumor Volume (mm3)')
TVPivoted_df = pd.DataFrame(TVtrans)
TVPivoted_df
```

```python
# Generate the Plot (with Error Bars)
cols = TVPivoted_df.columns[:]

for i in cols:
    plt.scatter(TVPivoted_df.index,TVPivoted_df[i])
    plt.errorbar(TVPivoted_df.index,TVPivoted_df[i],TVsemPivoted_df[i],linestyle="dashed",markersize="2")

plt.title('Tumor Response to treatment')
plt.xlabel("Timepoint (days)")
plt.ylabel("Tumor Volume (mm3)")
plt.legend(cols, loc='center left',bbox_to_anchor=(1, .5))
plt.savefig("Tumor Response to treatment.jpg",bbox_inches = "tight")
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
mean_metsites = pd.DataFrame(mouse_clinicaltrail_df.groupby(["Drug","Timepoint"])["Metastatic Sites"].mean())
mean_metsites.head()
```

```python
MetSites = mean_metsites.pivot_table(columns='Drug', index='Timepoint', values='Metastatic Sites')
MSPivoted_df = pd.DataFrame(MetSites)
MSPivoted_df
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
MetSites_sem_df = pd.DataFrame(mouse_clinicaltrail_df.groupby(["Drug","Timepoint"])["Metastatic Sites"].sem())
MetSites_sem_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
MSsemtrans = MetSites_sem_df.pivot_table(columns='Drug', index='Timepoint', values='Metastatic Sites')
MSsemPivoted_df = pd.DataFrame(MSsemtrans)
MSsemPivoted_df

# Preview that Reformatting worked
```

```python
for i in cols:
    plt.scatter(MSPivoted_df.index,MSPivoted_df[i])
    plt.errorbar(MSPivoted_df.index,MSPivoted_df[i],MSsemPivoted_df[i],linestyle="dashed",markersize="2")

plt.title('Metastatic spread during treatment')
plt.xlabel("Timepoint (days)")
plt.ylabel("Met. sites")
plt.legend(cols, loc='center left',bbox_to_anchor=(1, .5))
plt.savefig("Metastatic spread during treatment.jpg",bbox_inches = "tight")
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
SurvivalRate_df = pd.DataFrame(mouse_clinicaltrail_df.groupby(["Drug","Timepoint"])["Mouse ID"].count())
SurvivalRate_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
SRtrans = SurvivalRate_df.pivot_table(columns='Drug', index='Timepoint', values='Mouse ID')
SRPivoted_df = pd.DataFrame(SRtrans)
SRPivoted_df
# Preview the Data Frame
```

```python

SRPivotedPer_df = (SRPivoted_df/SRPivoted_df.max()*100)
SRPivotedPer_df
```

```python
# Generate the Plot (Accounting for percentages)
for i in cols:
    plt.plot(SRPivotedPer_df.index,SRPivotedPer_df[i],linestyle="dashed",marker='.')
    
plt.title('Survival rate suring treatment')
plt.xlabel("Timepoint (days)")
plt.ylabel("Survival Rate (%)")
plt.legend(cols, loc='center left',bbox_to_anchor=(1, .5))
plt.savefig("Survival rates.jpg",bbox_inches = "tight")
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
TVPerList = []

for i in cols:
    #mean_tumor_cap = mean_tumor.loc[mean_tumor["Drug"] == "Capomulin", :]
    TVlist = mean_tumor.filter(like=i, axis=0)['Tumor Volume (mm3)']
    TVPer = ((TVlist[len(TVlist)-1]/TVlist[0]*100)-100)
    TVPerList.append(TVPer)
    
# TVPerList = list(map(int, TVPerList))
TVSummary_df = pd.DataFrame(
     {'Drug':cols,'Percent Change':TVPerList})
TVSummary_df
```

```python
# Store all Relevant Percent Changes into a Tuple
locs = np.arange(len(cols)) + 0.5

plt.figure(figsize=(10,4))
for i in range(len(TVPerList)):
    if TVPerList[i]>0:
        plt.bar(cols[i],TVPerList[i],color='r',alpha=.75,align="center", width=1,edgecolor ="b")
        txtlable = str(int(TVPerList[i])) + "%"
        plt.text(cols[i],TVPerList[i],txtlable,ha='center', va='top',color='w',fontsize=10)
    else:
        plt.bar(cols[i],TVPerList[i],color='g',alpha=.75,align="center", width = 1,edgecolor ="b")
        txtlable = str(int(TVPerList[i])) + "%"
#         x_loc = cols[i] + 0.5
        plt.text(cols[i],TVPerList[i],txtlable,ha='center', va='bottom', color = 'w',fontsize=10)

plt.title('Tumor change over 45 days of treatment')
plt.yticks(fontsize=12)
plt.ylabel("% of Tumor volume change", fontsize=12)
plt.xticks(locs,cols, fontsize=12, rotation=90)
plt.grid()
plt.show()
plt.savefig("Tumor change over 45 days of treatment.jpg",bbox_inches = "tight")
```
