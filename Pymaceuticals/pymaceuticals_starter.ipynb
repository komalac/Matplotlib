```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy import stats

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_data = pd.read_csv(mouse_drug_data_to_load)
clinical_trial_data = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
# merge both datasets 
mouse_clinicaltrail_df = pd.merge(mouse_drug_data, clinical_trial_data, on = "Mouse ID")

# Display the data table for preview
mouse_clinicaltrail_df.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
mean_tumor = pd.DataFrame(mouse_clinicaltrail_df.groupby(["Drug","Timepoint"])["Tumor Volume (mm3)"].mean())
mean_tumor.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
TVsem_df = pd.DataFrame(mouse_clinicaltrail_df.groupby(["Drug","Timepoint"])["Tumor Volume (mm3)"].sem())
TVsem_df.head()
# Convert to DataFrame

# Preview DataFrame

TVsemtrans = TVsem_df.pivot_table(columns='Drug', index='Timepoint', values='Tumor Volume (mm3)')
TVsemPivoted_df = pd.DataFrame(TVsemtrans)
TVsemPivoted_df
```

```python
# Minor Data Munging to Re-Format the Data Frames

TVtrans = mean_tumor.pivot_table(columns='Drug', index='Timepoint', values='Tumor Volume (mm3)')
TVPivoted_df = pd.DataFrame(TVtrans)
TVPivoted_df
```

```python
# Generate the Plot (with Error Bars)
cols = TVPivoted_df.columns[:]

for i in cols:
    plt.scatter(TVPivoted_df.index,TVPivoted_df[i])
    plt.errorbar(TVPivoted_df.index,TVPivoted_df[i],TVsemPivoted_df[i],linestyle="dashed",markersize="2")

plt.title('Tumor Response to treatment')
plt.xlabel("Timepoint (days)")
plt.ylabel("Tumor Volume (mm3)")
plt.legend(cols, loc='center left',bbox_to_anchor=(1, .5))
plt.savefig("Tumor Response to treatment.jpg",bbox_inches = "tight")
```
